# Bakery Sales Dataset Analysis

*Plots*

```python
#Average Oil Price Over Time (2016-2017)
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('bakery_sales_dataset_4.5.csv')

df['Date'] = pd.to_datetime(df['Date'])

data_filtered = df[df['Date'].dt.year.isin([2016, 2017])]

plt.figure(figsize=(12,8))

plt.plot(data_filtered['Date'], data_filtered['Average Price'], label = 'Average Price', color='blue', marker='x')
plt.plot(data_filtered['Date'], data_filtered['Brent Price'], label = 'Brent Price', color='red')
plt.plot(data_filtered['Date'], data_filtered['WTI Price'], label = 'WTI Price', color='green')

plt.title('Average Oil Price Over Time (2016-2017)', fontsize=16)
plt.xlabel('Date')
plt.ylabel('Average Oil Price')

plt.grid(True)
plt.legend()
plt.show()
```

```python
#Weather Data Over Time (2016-2017)
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('bakery_sales_dataset_4.5.csv')

df['Date'] = pd.to_datetime(df['Date'])
df['pres'] = df['pres'] / 1000 #mbar to bar

data_filtered = df[df['Date'].dt.year.isin([2016, 2017])]

plt.figure(figsize=(12,8))
plt.title('Weather Data Over Time (2016-2017)', fontsize=16)
plt.xlabel('Date')
plt.ylabel('Weather Data')

plt.plot(data_filtered['Date'], data_filtered['tmax'], label='Maximum Temprature', color='blue')
plt.plot(data_filtered['Date'], data_filtered['tmin'], label='Minimum Temprature', color='yellow')
plt.plot(data_filtered['Date'], data_filtered['tavg'], label='Average Temprature', color='green')
plt.plot(data_filtered['Date'], data_filtered['wspd'], label='Wind Speed', color='red')
plt.plot(data_filtered['Date'], data_filtered['pres'], label='Pressure (bar)', color='orange')

plt.legend()
plt.show()
```

```python
#Cumulative Demands Of Items
import matplotlib.pyplot as plt

plt.figure(figsize=(12,8))

names = []
values = []
with open('unique_items_ip.txt', 'r') as f:
    for line in f:
        name, value = line.strip().split()
        names.append(name)
        values.append(int(value))

bars = plt.bar(names, values, color='navy')

#Creating histogram
for i, bar in enumerate(bars):
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width() / 2, height + 100,
             f'{names[i]}', ha='center', va='bottom', fontsize=8,
             color='black', rotation=90)

plt.title('Cumulative Demands Of Items')
plt.xlabel('Items')
plt.ylabel('Counts')
plt.xticks([])
plt.ylim(0, 6000)

plt.show()
```

*Corelation* *Analysis*

```python
#Corealtion Analysis of Prices
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv('bakery_sales_dataset_4.5.csv')
columns = ['Average Price', 'Brent Price', 'WTI Price']

matrix = df[columns].corr()

sns.heatmap(matrix, annot=True, cmap='coolwarm', fmt='.2f', linewidths=0.5)

plt.title('Correlation Matrix Of Prices')
plt.figure(figsize=(12,8))

plt.show()
```

```python
#Corealtion Analysis of Items
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv('bakery_sales_dataset_4.5.csv')
columns = ['filter coffee', 'bread', 'croissant', 'angbutter', 'tea', 'cake', 'cookie',
                        'plain bread', 'pastry', 'sandwich', 'hot chocolate', 'orange pound', 'latte',
                        'americano', 'cacao deep', 'wiener', 'pandoro', 'brownie', 'jam', 'farm house bread',
                        'juice', 'muffin', 'scone', 'soup', 'toast', 'scandinavian', 'gateau chocolat',
                        'truffles', 'coke', 'spanish brunch', 'fudge', 'baguette bread', 'tiffin', 'milk tea',
                        'chicken stew', 'jammie dodgers', 'mineral water', 'hearty & seasonal', 'cheesecake',
                        'salad', 'frittata', 'smoothies', 'the nomad', 'vegan mincepie', 'berry ade', 'focaccia']

matrix = df[columns].corr()

sns.heatmap(matrix, annot=True, cmap='coolwarm', fmt='.2f', linewidths=0.5,
            cbar_kws={'shrink': 0.75}, annot_kws={'size': 8}, xticklabels=1, yticklabels=1)

plt.xticks(rotation=90)
plt.yticks(rotation=0)

plt.title('Correlation Matrix Of Weather Values')
plt.figure(figsize=(20,16))

plt.show()
```
